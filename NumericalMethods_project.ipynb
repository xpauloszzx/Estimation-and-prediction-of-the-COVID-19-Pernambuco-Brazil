<h1> Métodos </h1>

A metodologia que vamos propor trata-se de  reformular um modelo matemático, baseado no modelo SIR [[0]](https://pt.wikipedia.org/wiki/Modelo_epid%C3%AAmico) (modelo epidemiológico de descrição de pessoas suscetíveis, infectados e recuperados) de Kermack Mackendrick dada uma população fechada (habitantes de Pernambuco), que  possa descrever a progressão da COVID-19 no período de (01/01/2022 - 31/07/2022) no estado de Pernambuco, como já anteriormente citado. O modelo a ser utilizado será enriquecido por mais três compartimentos: questões relacionadas a vacinação da população, a exposição ao vírus e os falecidos por consequência da doença. Diversos parâmetros serão considerados para tornar válida a construção do modelo. Para  a questão da organização dos dados e previsão, será utilizado python, juntamente com a sua API SciPy.integrate para resolução do conjunto de equações diferenciais, que irá prever e descrever o futuro comportamento do vírus, também  com a API matplotlib e numpy para a análise exploratória de dados e plotagem gráfica. Em detalhe, no modelo epidemiológico será considerado tais compartimentos para a modelagem: 



**Suscetíveis[S]:** É um grupo da população que não está infectado, mas possui riscos de contrair a doença, seja este vacinado ou não. Como a Covid se trata de um vírus de alto contágio, doenças causadas por esse tipo de agente necessitam de contato entre o hospedeiro(carregando o vírus) e um indivíduo sucetível. Dado exposto, é necessário considerarmos que os indivíduos suscetíveis que contrair a doença serão convertidos para infectados, possivelmente para recuperados(não se torna imune ao vírus) ou mortos. Por consequência, temos alguns pararâmetros definidos como: 

**(S0)** - Estágio inicial de susceptibilidade: Os indivíduos suscetíveis iniciam  em um estágio R0 e podem ser convertidos para diferentes estágios da doença ao longo de um período fechado t (01/01/2022 - 31/07/2022);

**(S)** - Grupo de indivíduos suscetíveis;

**Exposição[E(t)]:** Trata-se de pessoas que estão expostas ao vírus, em ambientes insalubres: organizacional, nos transportes coletivos, hospitais, concertos, mercado, etc. É relacionado a pessoas que não são imune ao vírus e estão sempre em possibilidade de transmitir/ser infectadas. Ademais, é um parâmetro que depende signficamente dos métodos de prevenção do usuário: higienização, distanciamento social, pouco tempo de exposição em lugares superlotados. De certa forma, em uma distância de 1 metro, se houver um hospedeiro do agente viral, apenas 15 minutos são necessários para a
contaminação, segundo a Organização Mundial de Saúde (OMS)[[1]](https://www.prefeitura.sp.gov.br/cidade/secretarias/saude/vigilancia_em_saude/doencas_e_agravos/coronavirus/index.php?p=295099); Com isso, tem-se novos parâmetros: 

**(d)** - Tempo de incubação: Se refere ao tempo de exposição do vírus até o desenvolvimento dos sintomas, entre 2-14 dias [[2]](https://www.worldometers.info/coronavirus/coronavirus-incubation-period/).

**(E)** - Grupo de expostos;

**Infectados[I(t)]**: Se refere ao grupo de pessoas afetadas pelo vírus, seja um grupo fechado de indivíduos já imunizados com a vacinação ou não. É considerado tanto os casos graves, assintomáticos e os de grau mais leve. Em muitos casos, o número de infectados pode ser acrescentado dependendo do tempo de exposição t em relação a um hospedeiro que contém o vírus.Com efeito, segundo dados da Secretaria  Estadual de saúde de pernambuco, cerca de 885 mil pessoas foram infectadas pelo vírus em Pernambuco (até a segunda metade do mês de março de 2022)[[3]](https://dados.seplag.pe.gov.br/apps/corona.html). Além disso, é afirmado, pelo mesmo orgão, que pelo menos 2.000 novos casos são acrescentados a estes números todos os dias e que a taxa de infecção do vírus (durante o mês de janeiro de 2022 até então) se iguala a 2,02 [[4]](https://g1.globo.com/saude/coronavirus/noticia/2022/02/01/taxa-de-transmissao-do-coronavirus-cai-para-169-no-brasil-diz-imperial-college.ghtml), muito menor que no auge da pandemia, onde os números se igualavam a 1,01 [[5]](https://agenciabrasil.ebc.com.br/saude/noticia/2020-08/taxa-de-transmissao-da-covid-19-no-brasil-cai-para-101-diz-estudo).

**(I)** - Grupo de indivíduos infectados;

**(α)** - Taxa de incubação; 

**(β)** - Taxa de infecção;

**Recuperados[R(t)]**: Trata-se do grupo de pessoas que estão totalmente curadas do vírus, porém ainda com chances de reinfecção, dado  pelo parâmetro **(γ)**.Pesquisas realizadas pela secretaria de saúde de Pernambuco apontam que a taxa de recuperação da população pernambucana ultrapassa os 83% de casos confirmados, com 734.426 recuperados de 882.753 casos confirmados(dados de março de 2022) [[3]](https://dados.seplag.pe.gov.br/apps/corona.html). Define-se os parâmetros: 

**(R)** - Grupo de recuperados;
**(R)** - Número inicial de recuperados;
**(γ)** - Taxa de reinfecções;

**Vacinados[V(t)]**: É um compartimento que influencia diretamente na taxa de recuperados e está intrísecamente ligada ao número de casos fatais, visto que a vacinação dificulta a aderência da covid-19 em seu estado grave, o que resulta em uma recuperação mais fácil e menor taxa de letalidade. A Organização Mundial de Saúde, através de suas pesquisas estatísticas, por exemplo, comprova que entre os vacinados, que, embora possam se infectar novamente, desenvolverão, na grande maioria dos casos, um quadro moderado; a segunda, entre os não vacinados, representam entre 80% e 90% dos pacientes com infecções graves, internações e óbitos [[4]](https://www.saude.mg.gov.br/component/gmg/story/16134-vacina-diminui-risco-de-complicacao-de-covid-19#:~:text=Dados%20do%20Minist%C3%A9rio%20da%20Sa%C3%BAde,19%20%C3%A9%20de%20apenas%200%2C005%25.). Além disso, é afirmado que 17,5 milhões de doses já foram aplicadas nos municípios de Pernambuco, e que aproximadamente 2,9 milhões de doses de reforço e pelo menos  6 e 8 milhões de primeira e segunda dose, respectivamente [[5]](https://drive.google.com/file/d/1NhDZzljZX2iSH7uh8AtHltdbuxxSbdo4/view?fbclid=IwAR1r9G-marrT3CPTEPGT2klpxU4ltG6OkMQTD1JhuGiYTZICib3p1lcaS5w).   

**(V)** - Quantidade de indivíduos vacinados;

**(υ)** - Taxa imunização. Será considerado valores entre [0,1], em respeito a quantidade de doses tomadas por indivíduo. Sendo a 1º dose: um valor variável entre 0 < υ < 0,5, a 2º dose: 0,5 < υ < 0,8, a terceira dose 0,8< υ < 1. É considerado que a vacinação não  torna o índividuo totalmente imune ao vírus, mas diminui significativamente a sua aderência em casos mais graves, buscando diminuir o número de casos removíveis. 

**Mortos[D(t)]**: É responsável por quantificar e classificar o número de falecidos que foram acometidos pelo Covid-19. Até o mês de março de 2022, a Secretaria de Saúde (SES) afirma que cerca de 21.305 casos de 882.753 infectados foram óbitos. **(η)** - Fração dos indivíduos falecidos(removíveis);

**(D)** - Quantidade de casos removíveis;

**(ψ)** - Taxa de mortalidade;

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#SEIRD
def SEIRD(A, t,alpha, beta, gamma, N, sigma, l4mbda, xi):
    #definindo as variáveis.
    S,E,I,R,D=A
    return [
        (-beta/N) * S * I,
        ((beta * S * I)/N) - sigma * E,
        (sigma * E) - ((1-alpha)*gamma*I) - (alpha*l4mbda*I),
        ((1-alpha)*gamma*I)+(alpha*(1-xi)*l4mbda*I),
        alpha * xi * l4mbda*I
    ]

#Quanto tempo será simulado
times = np.arange(0, 100, 1)

# ctes
sigma=0.192
beta=0.05*16
alpha=0.2
xi=0.3
l4mbda=0.143
gamma=0.313

#Valores iniciais das variáveis
N=1e7
S0, E0, I0, R0, D0 = N-574, 574, 1, 0, 0

#resolvendo a ODE.
sol=odeint(SEIRD, y0=[S0, E0, I0, R0, D0],t=times, args=(alpha,beta,gamma,N,sigma,l4mbda,xi))

#plotando os gráficos.
S,E,I,R,D=sol.T

plt.figure(facecolor='w',figsize=(30,10))
plt.plot(times, S,label='Suscetíveis')
plt.plot(times, E,label='Expostos')
plt.plot(times, I,label='Infectados')
plt.plot(times, R,label='Recuperados')
plt.plot(times, D,label='Mortos')
plt.legend() 
plt.title("Progressão da COVID-19 no estado de Pernambuco no momento de pré-vacinação:")
plt.xlabel(("Tempo(Dias)"))
plt.ylabel(("Porcentagem de indivíduos(%)"))

plt.grid()
plt.show()



#SEIRDV

def dAdt(A, t, beta, gamma, N, nu, alpha, eta):
    #definindo as variáveis.
    S = A[0]
    E = A[1]
    I = A[2]
    R = A[3]
    D = A[4]
    V = A[5]
    return [
        -beta/N * S * I - v*S,
        beta/N * S * I - alpha * E,
        alpha * E - gamma * I,
        gamma * (1 - eta) * I,
        gamma * eta * I,
        v*S
    ]

#Quanto tempo será simulado
times = np.arange(0, 500, 1)

#definindo os valores das constantes
gamma = 0.083
N = 1e7
beta = 0.20
v = 0.0039
alpha = 0.192
eta = 0.024

#Valores iniciais das variáveis
S0, E0, I0, R0, D0, V0 = N-800, 800, 0, 0, 0, 0

#resolvendo a ODE.
sol = odeint(dAdt, y0=[S0, E0, I0, R0, D0, V0],t=times, args=(beta, gamma, N, v, alpha, eta))

#plotando os gráficos.
S = sol.T[0]
E = sol.T[1]
I = sol.T[2]
R = sol.T[3]
D = sol.T[4]
V = sol.T[5]

plt.figure(facecolor='w',figsize=(30,10))
plt.plot(times, S,label='Suscetíveis')
plt.plot(times, E,label='Expostos')
plt.plot(times, I,label='Infectados')
plt.plot(times, R,label='Recuperados')
plt.plot(times, D,label='Mortos')
plt.plot(times, V,label='Vacinados')
plt.title("Progressão da COVID-19 no estado de Pernambuco após a 2º dose:")
plt.xlabel(("Tempo(Dias)"))
plt.ylabel(("Porcentagem de indivíduos(%)"))
plt.legend()
plt.grid()
plt.show()

#SEIRDV

def dAdt(A, t, beta, gamma, N, nu, alpha, eta):
    #definindo as variáveis.
    S = A[0]
    E = A[1]
    I = A[2]
    R = A[3]
    D = A[4]
    V = A[5]
    return [
        -beta/N * S * I - v*S,
        beta/N * S * I - alpha * E,
        alpha * E - gamma * I,
        gamma * (1 - eta) * I,
        gamma * eta * I,
        v*S
    ]

#Quanto tempo será simulado
times = np.arange(0, 500, 1)

#definindo os valores das constantes
gamma = 0.083
N = 1e7
beta = 0.20
v = 0.06
alpha = 0.192
eta = 0.024

#Valores iniciais das variáveis
S0, E0, I0, R0, D0, V0 = N-800, 800, 0, 0, 0, 0

#resolvendo a ODE.
sol = odeint(dAdt, y0=[S0, E0, I0, R0, D0, V0],t=times, args=(beta, gamma, N, v, alpha, eta))

#plotando os gráficos.
S = sol.T[0]
E = sol.T[1]
I = sol.T[2]
R = sol.T[3]
D = sol.T[4]
V = sol.T[5]

plt.figure(facecolor='w',figsize=(30,10))
plt.plot(times, S,label='Suscetíveis')
plt.plot(times, E,label='Expostos')
plt.plot(times, I,label='Infectados')
plt.plot(times, R,label='Recuperados')
plt.plot(times, D,label='Mortos')
plt.plot(times, V,label='Vacinados')
plt.title("Progressão da COVID-19 no estado de Pernambuco após a dose de reforço (3º dose):")
plt.xlabel(("Tempo(Dias)"))
plt.ylabel(("Porcentagem de indivíduos(%)"))
plt.legend()
plt.grid()
plt.show()

```

Assim sendo, após  uma grande taxa de indivíduos vacinados com aplicação das 3 doses propostas, há grandes indícios que acontecerá a atenuação da transmissão em seu estado natural, visto que um dos principais  objetivos da vacinação consiste em diminuir significativamente o número de casos graves e por conseguinte, o número de casos removíveis. Esse fato pode ser comprovado por uma pesquisa conduzida em Londrina, no Paraná, pela Universidade Estadual de Londrina e da Faculdade de Medicina Albert Einstein dos Estados Unidos, que mostrou que 75% das mortes por Covid-19 registradas nos primeiros dez meses de 2021 (período entre as primeiras ondas) ocorreram em indivíduos que não foram imunizados contra a doença. Os idosos não vacinados morreram quase três vezes mais do que os imunizados. Entre pessoas com menos de 60 anos, o número de mortes de não vacinados foi 83 vezes maior do que nos imunizados [[8]](https://butantan.gov.br/noticias/nao-vacinados-representam-75-das-mortes-por-covid-19-diz-estudo-brasileiro
  ).

 Além disso, é importante frisar que o distanciamento social foi outro fator importante para diminuir a transmissão da COVID-19, visto que é uma recomendação médica  que consiste um conjunto de ações que buscam limitar o convívio social de modo a parar ou controlar a propagação de doenças contagiosas. Seu objetivo é reduzir a probabilidade de contato entre pessoas portadoras de uma infecção com outras que não estão infectadas, minimizando a transmissão da doença e, consequentemente, a mortalidade. No nosso modelo utilizado, a questão do distanciamento social está intrísecamente ligado ao parâmetro *β* (Beta)- que será explicitado com mais detalhes na seção da obtenção dos parâmetros do modelo SEIRDV-, que depende de mais dois outros fatores: a probabilidade **p** de um indivíduo infectar o outro caso haja contato entre os dois e **n**, que se trata de uma média  de encontro dos indivíduos por dia. Nesse sentido, assim como visto nas equações diferenciais, o parâmetro **β** afeta diretamente nas variáveis relativas ao número de infectados (I) e expostos (E). Com isso, percebe-se que o distanciamento social, assim como a vacinação foi umas das medidas mais importantes e eficazes para reduzir o avanço da pandemia da covid-19, visto que altera consideravelmente os  números de casos de infectados e a exposição dos cidadãos aos vírus. Esse fato é explictado através de um estudo realizado pela Universidade Federal Rural do Rio de Janeiro, observando o comportamento do vírus da COVID-19 no ano de 2020, baseando-se em dados de comportamento da pandemia no país, taxa de isolamento e diferenças regionais. Os estudiosos estimaram que a cada 1% de aumento no isolamento social  existiu uma redução de até 37% na taxa de propagação do vírus [[9]](https://www.sanarmed.com/fake-news-isolamento-social-e-ineficaz-e-80-da-populacao-e-imune-a-covid-19).

 


<h3> Comparação da Segunda e terceira onda </h3>

<p> A segunda onda foi caracterizada principalmente pela presença das variantes delta e gamma, a predominância de uma variante se dá pela seu maior poder de contaminação e pela maior possibilidade do indivíduo desenvolver casos mais graves da doença, com isso, torna-se evidente que, após o surgimento de uma variante, deve-se atentar para medidas mais rígidas de combate, sendo uma delas a vacinação, que impede, na maioria da vezes, o indivíduo de desenvolver o estágio mais grave. Sendo assim, a campanha de vacinação com 2 doses foi uma importante ferramente de combate a pandemia, pois certamente evitou a superlotação do sistema público de saúde e diminui o número de óbitos. Já na terceira onda do covid em Pernambuco, estávamos diante de uma variante ainda mais poderosa, a omicron. No entanto, a terceira onda foi mais curta em comparação com a segunda,  visto que uma porcentagem considerável da população mais vulnerável já se encontrava vacinada com a dose de reforço (3 doses), o que, seguindo a simulação do terceiro gráfico, resultou em um menor número de casos e de óbitos.</p>

<h1> Obtenção dos parâmetros SEIRDV </h1>

β está relacionado ao número de contatos entre Suscetíveis e Infectados. Em termos gerais, β depende de outros dois parâmetros: p que é a probabilidade de um indivíduo I infectar uma pessoa S e n, que é uma média de encontros dos indivíduos (I com S) por dia. 

$β=pn$

p e n são duas variáveis que dependem de muitos fatores, que foram afetas (e ainda são) ao longo do tempo pelo uso de máscaras, higienização, respeito a quarentena, entre outros.

Para nosso projeto, achamos válido definir p e n como constantes, tentando tirar uma média de tais variáveis, os resultados condirados foram:

$p = 2\%$ e $n = 10$

Dessa forma, 

\begin{align}
        β=0,02*10=0,2
    \end{align}



O período de incubação (tempo desde a exposição ao desenvolvimento dos sintomas) do COVID-19 varia bastante, segundo a Organização Mundial da Saúde (OMS) esse número pode oscilar entre 2 e 10 dias. Em um estudo chinês publicado no New England Journal of Medicine em 29 de janeiro de 2020, foi relatado que a taxa média do período de incubação é de 5,2 dias.

Se consideramos que o período de incubação pode ser aproximado por uma distribuicão exponencial com média α, isto é:
\begin{align}
        T_{Minc} ≈ exp(α) ⇒ α = \frac{1}{T_{Minc}}=\frac{1}{5,2}≈0,192
    \end{align}

Estudos revelaram que o período de transmissão do vírus começa por volta de dois dias antes do surgimento do primeiro sintoma, durando até 10 dias após a aparição deles. Consideramos então que uma pessoa permanece infectante por 12 dias em média. Novamente assumimos que esse período pode ser aproximado por uma distribuição exponencial, ou seja

\begin{align}
        T_{Mi} ≈ exp(γ) ⇒ γ = \frac{1}{T_{Mi}}=\frac{1}{12}≈0,083
    \end{align}



Após esse período $T_{Mi}$, os Infectados se dividem em Recuperados e Mortos com uma taxa de 1-η e η, respectivamente.

De acordo com o último boletim (até a realização do projeto) da Secretaria Estadual de Saúde (SES-PE), divulgado em 25/03, o estado registrou um total de 889000 casos confirmados da doença e 21345 mortes. Portanto, a taxa de mortes é de

\begin{align}
        η = \frac{21345}{889000}≈0,024
    \end{align}

# Conclusão

Portanto, propusemos um modelo SEIRDV para análise da progressão da COVID-19 durante um período do ano de 2021 e 2022 para o estado de Pernambuco e SEIRD para análise de 2020. Os resultados sobre os dados de todo estado de Pernambuco são muito promissores, visto que podemos perceber uma menor quantidade de mortes e de infectados. 

As previsões simuladas do comportamento da população infectada no período de 2022 mostram que o modelo tem um melhor desempenho quando comparado com os dados disponíveis nesse período. Em particular, no caso do estado de Pernambuco, observamos que o dia de pico para a população infectada é previsto com grande precisão.

Simulações da curva infectada com uma taxa de vacinação diferente e crescente no mesmo período dão a ideia de como a epidemia iria evoluir nestes casos. Os vacinados podem se infectar novamente, mas na maioria dos casos vão ter um quadro moderado. Agora os não vacinados vão representar 80% e 90% dos pacientes com infecções graves, internações e óbitos.

Estudos futuros e extensões deste modelo irão considerar a duração limitada da imunidade induzida pela vacina e o possível padrão sazonal das ondas epidêmicas da COVID-19.
